### 1. Importing and exploring the DataFrame

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_amazon= pd.read_csv('../data/raw/Amazon.csv')
df_amazon.head()
```

```python
df_amazon.shape
```

Columns information:

- **index** Internal row index imported by pandas; not part of the actual business data. 
- **Order ID** Unique identifier for each order or transaction. 
- **Date** Date the order was placed. Often stored as text and should be converted to `datetime`. 
- **Status** Current order status (e.g., Delivered, Cancelled, Returned). 
- **Fulfilment** Fulfilment method (Amazon vs. third-party seller). 
- **Sales Channel** Platform or channel where the sale occurred (e.g., Amazon marketplace, mobile app). 
- **ship-service-level** Shipping service level selected (standard, expedited, priority, etc.). 
- **Style** Product style or variant attribute (e.g., design, color). 
- **SKU** Stock Keeping Unit; internal product code for inventory tracking. 
- **Category** Product category or classification. 
- **Size** Size variant of the product. 
- **ASIN** Amazon Standard Identification Number; unique product identifier on Amazon. 
- **Courier Status** Shipment status from the courier’s perspective (e.g., in transit, delivered, delayed). 
- **Qty** Quantity of units ordered in the order line. 
- **currency** Currency code used for the transaction (e.g., USD, INR). 
- **Amount** Monetary value of the order line (total revenue for that line). 
- **ship-city** Destination city for shipping. 
- **ship-state** Destination state or region for shipping. 
- **ship-postal-code** Postal code of the shipping address. 
- **ship-country** Destination country for shipping. 
- **promotion-ids** Identifiers of applied promotions or discount campaigns, if any. 
- **B2B** Indicates whether the order is Business-to-Business (B2B) or retail. 
- **fulfilled-by** Entity responsible for fulfilment (Amazon, third-party seller, logistics partner). 
- **Unnamed: 22** Empty or unused column; CSV artifact with no data. 

### 2. Cleaning names of columns

```python
df_amazon.columns = (
    df_amazon.columns
      .str.lower()
      .str.normalize('NFKD')      
      .str.encode('ascii', errors='ignore')
      .str.decode('utf-8')
      .str.replace(' ', '_')
      .str.replace('[^0-9a-zA-Z_]', '')
)
df_amazon.head(5)
```
